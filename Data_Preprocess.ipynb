```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python


# Load the loan data into a pandas dataframe
df = pd.read_csv('hw3.csv')
df

```

### Data Preprocessing

```python
df['grade'].value_counts()
```

```python
df['loan_status'].value_counts( )
```

```python
df['home_ownership'].value_counts()
```

```python
df['delinq_2yrs'].value_counts()
```

```python
# Remove the index column
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df

```

```python
# Replace missing values with the mean value of the variable
df.fillna(df.mean(), inplace=True)
df
```

```python
# Encode categorical variables using one-hot encoding
df = pd.get_dummies(df, columns=['home_ownership', 'grade', 'term'])
df
```

```python
# Label encode the target variable
le = LabelEncoder()
df['loan_status'] = le.fit_transform(df['loan_status'])
df
```

```python
# Scale the numerical variables using standardization
scaler = StandardScaler()
numerical_cols = ['annual_inc', 'delinq_2yrs', 'dti', 'revol_bal', 'total_acc', 'funded_amnt', 'total_pymnt']
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
df

```
